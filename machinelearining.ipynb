```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
```

```python
df = pd.read_csv(r"C:\Users\samud\Downloads\loan.csv.csv")
df
```

```python
df.head (4)
```

```python
df.tail(4)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.select_dtypes(include ="float64").columns
```

```python
# by using the sklearn imputer we can fill the dataset 
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy='mean') # mean we can select beacuse numerical value are there
```

```python
si.fit_transform(df[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']]) # data is fill by array in form
```

```python
# but we can transfrom the data array to dataset 
```

```python
ar =si.fit_transform(df[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

```python
di = pd.DataFrame(ar,columns =df.select_dtypes(include ="float64").columns)
di
```

```python
di.isnull().sum()
```

```python
sns.heatmap(di.isnull())
```
